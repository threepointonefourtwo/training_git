```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model, tree, model_selection
```

```python
df = pd.read_csv("train.csv")
#first plot
plt.subplot(2,2,1)
plt.plot(df.Survived,df.Age,'y')
plt.xlabel("Survived")
plt.ylabel("age")
#second plot
plt.subplot(2,2,2)
df.Survived.value_counts().plot(kind = "bar")
plt.xlabel("Survived")
plt.ylabel("age")
#third plot
plt.subplot(2,2,3)
for x in [1,2,3]:
    df.Age[df.Pclass == x].plot(kind = "kde")
plt.legend(("1st","2nd","3rd"))
#fourth plot
plt.subplot(2,2,4)
df.Embarked.value_counts().plot(kind="bar")
plt.show()
```

```python
plt.subplot(2,3,1)
df.Survived[df.Sex == "male"].value_counts().plot(kind = "bar")
plt.xlabel("Male Survived")
plt.ylabel("age")
plt.subplot(2,3,2)
df.Survived[df.Sex == "female"].value_counts().plot(kind = "bar")
plt.xlabel("Female Survived")
plt.ylabel("age")
plt.subplot(2,3,3)
df.Sex[df.Survived == 1].value_counts().plot(kind = "bar")
plt.ylabel("age")
plt.subplot(2,3,4)
for x in [1,2,3]:
    df.Survived[df.Pclass == x].plot(kind = "kde")
plt.legend(("1st","2nd","3rd"))
plt.tight_layout()
plt.show()
```

```python
plt.subplot(2,2,1)
df.Survived[(df.Sex == "male") & (df.Pclass == 1)].value_counts().plot(kind = "bar")
plt.title("Rich Men Survived")
plt.subplot(2,2,2)
df.Survived[(df.Sex == "male") & (df.Pclass == 3)].value_counts().plot(kind = "bar")
plt.title("Poor Men Survived")
plt.subplot(2,2,3)
df.Survived[(df.Sex == "female") & (df.Pclass == 1)].value_counts().plot(kind = "bar")
plt.title("Rich Women Survived")
plt.subplot(2,2,4)
df.Survived[(df.Sex == "male") & (df.Pclass == 1)].value_counts().plot(kind = "bar")
plt.title("poor Women Survived")
plt.tight_layout()
```
